```python
li = [5,6,1,3,2]
n = len(li)
for i in range(n-1):
    ifswapp = False
    for j in range(n-i-1):
        if li[j] > li[j+1]:
            li[j],li[j+1] = li[j+1], li[j]
            ifswapp = True
    print(li)
    if (ifswapp == False):
        break
print(li)
```
