```python
rows = 4
col = 4

for i in range(rows):
    for j in range(col):
        print("*",end='')
        print('_',end='')
    print()
```

```python
rows = 4
col = 4

for i in range(rows):
    for j in range(col):
        print("*",end='')
        if j!= col-1:
            print('_',end='')
    print()
```
