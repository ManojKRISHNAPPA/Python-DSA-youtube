```python
colors = [1,1,1,6,1,1,1]

n = len(colors)

for i in range(n):
    for j in range(i+1,n):
        print(i,j)
```

```python
colors = [1,1,1,6,1,1,1]

n = len(colors)

for i in range(n):
    for j in range(i+1,n):
        print(colors[i],colors[j])
```

```python
colors = [1,1,1,6,1,1,1]

n = len(colors)

for i in range(n):
    for j in range(i+1,n):
        if colors[i] != colors[j]:
            print(colors[i],colors[j])
```

```python
colors = [1,1,1,6,1,1,1]

n = len(colors)
ans = 0
for i in range(n):
    for j in range(i+1,n):
        if colors[i] != colors[j]:
           temp = j-i
           ans = max(ans,temp)

print(ans)
```

```python
colors = [1,8,3,8,3]

n = len(colors)
ans = 0
for i in range(n):
    for j in range(i+1,n):
        if colors[i] != colors[j]:
           temp = j-i
           ans = max(ans,temp)

print(ans)
```

```python
colors = [1,1,1,6,1,1,1]
n = len(colors)
ans = 0
for i in range(n-1,-1,-1):
    if colors[0] != colors[i]:
        temp=i
        ans=max(temp,ans)
        break
print(ans)
```

```python
colors = [1,1,1,6,1,1,1]
n = len(colors)
ans = 0
for i in range(n-1,-1,-1):
    if colors[0] != colors[i]:
        temp=i
        ans=max(temp,ans)
        break
for i in range(n):
    if colors[i] != colors[n-1]:
        temp=n-1-i
        ans=max(temp,ans)
        break
print(ans)
```
