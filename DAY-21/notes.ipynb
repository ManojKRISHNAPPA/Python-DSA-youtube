```python
# Write a program to enter name and percentage marks in dictinory and disply the information
```

```python
rec={}
n=eval(input("enter number of students: "))
i=1
while i<=n:
    name=input("Enter student name: ")
    marks=eval(input("Enter marks: "))
    rec[name]=marks
    i=i+1
print((f"Name of the student is {name} and marks is {marks}"))

```

```python
print(rec)
```

```python
for x in rec:
    print(f'name is {x} and marks is {rec[x]}')
```

```python
li = [1,5,3,2,1,2,4,5,6,7,8,1,2,4]
n = len(li)
d={}
for i in range(n):
    val=li[i]
    if val not in d:
        d[val]=1
    else:
        temp = d[val]
        d[val]=temp+1
print(d)
```

```python
li = [1,1,1,1,1,1,5,5,5,2,2]
d={}
for val in li:
    if val not in d:
        d[val]=1
    else:
        d[val] = d[val]+1
print(d)
```
