```python
nums = [5,4,1,8]
n = len(nums)
for i in range(n):
    for j in range(i+1,n):
        print(i,j)
```

```python
nums = [5,4,1,8]
n = len(nums)
for i in range(n):
    for j in range(i+1,n):
        print(nums[i],nums[j])
```

```python
nums = [5,4,1,8]
n = len(nums)
for i in range(n):
    for j in range(i+1,n):
        if nums[i] < nums[j]:
            print(nums[i],nums[j])
```

```python
nums = [5,4,1,8]
n = len(nums)
ans = 0 
for i in range(n):
    for j in range(i+1,n):
        if nums[i] < nums[j]:
            temp = nums[j] - nums[i]
            ans = max(ans,temp)
print(ans)
```

```python
nums = [9,4,3,2]
n = len(nums)
ans = -1 
for i in range(n):
    for j in range(i+1,n):
        if nums[i] < nums[j]:
            temp = nums[j] - nums[i]
            ans = max(ans,temp)
print(ans)
```

```python
nums = [9,4,3,2]
n = len(nums)
ans = -1 
low = nums[0]
for j in range(i+1,n):
    if low < nums[j]:
        temp = nums[j] - low
        ans = max(ans,temp)
    low = min(low,nums[j])
print(ans)
```

```python
nums = [5,4,2,7,9]
n = len(nums)
ans = -1 
low = nums[0]
for j in range(1,n):
    if low < nums[j]:
        temp = nums[j] - low
        ans = max(ans,temp)
    low = min(low,nums[j])
print(ans)
```
