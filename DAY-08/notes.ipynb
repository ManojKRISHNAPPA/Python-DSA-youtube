### 1.https://leetcode.com/problems/final-value-of-variable-after-performing-operations/description/

There is a programming language with only four operations and one variable X:

++X and X++ increments the value of the variable X by 1.
--X and X-- decrements the value of the variable X by 1.
Initially, the value of X is 0.

Given an array of strings operations containing a list of operations, return the final value of X after performing all the operations.

Example 1:

Input: operations = ["--X","X++","X++"]
Output: 1
Explanation: The operations are performed as follows:
Initially, X = 0.
--X: X is decremented by 1, X =  0 - 1 = -1.
X++: X is incremented by 1, X = -1 + 1 =  0.
X++: X is incremented by 1, X =  0 + 1 =  1.


```python
operations = ["--X","X++","X++"]
n=len(operations)
X=0
for i in range(n):
    temp=operations[i]

    if temp == "--X":
        X=X-1
    if temp == "X--":
        X=X-1
    if temp = "++X":
        X=X+1
    if temp = "X++":
        X=X+1
print(X)
```

# Defanging an IP Address


Example 1:

Input: address = "1.1.1.1"
Output: "1[.]1[.]1[.]1"
Example 2:

Input: address = "255.100.50.0"
Output: "255[.]100[.]50[.]0"

```python
address = "1.1.1.1"
n = len(address)

ans=''
for i in range(n):
    ch = address[i]

    if ch == ".":
        ans = ans+'[.]'
    else:
        ans = ans+ch
print(ans)
```

Given the head of a sorted linked list, delete all duplicates such that each element appears only once. Return the linked list sorted as well.


```python
# Define a linked list node
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

# Helper function to create linked list from list
def create_linked_list(arr):
    dummy = ListNode(0)
    curr = dummy
    for val in arr:
        curr.next = ListNode(val)
        curr = curr.next
    return dummy.next

# Helper function to print linked list
def print_linked_list(head):
    res = []
    while head:
        res.append(head.val)
        head = head.next
    print(res)

# Input
head = create_linked_list([1, 1, 2])
res = head

# Remove duplicates
while head and head.next:
    if head.val == head.next.val:
        head.next = head.next.next
    else:
        head = head.next

# Output
print_linked_list(res)
```

## Approach

## 🧠 Removing Duplicates from a Sorted Linked List

### 🔹 Problem Example

**Input:**  

head = 1 → 1 → 2


All we have to do is **remove the next duplicate node** and connect the current node with the next-next node.

---

### ⭐️ Key Idea

When we find duplicate values,  
👉 **Connect the current node with the next-next node.**

---

### 🔍 Step-by-Step Example

#### Example 1
1 → 1 → 2
h

We found a duplicate (`1 == 1`), so:
head.next (= 1) = head.next.next (= 2)
Now the list becomes:
1 → 2
h

---

### ⚠️ Important Observation

We **should not move the head pointer** to the next node **when we find a duplicate**.

#### Why?

Consider this example:


1 → 1 → 1 → 2
h


We find duplicates (first `1` and second `1`), so we remove the second one:

If we move `head` to the next node now:
1 → 1 → 2
h

We miss checking the remaining duplicate because we’re already at the **second `1`**.

So instead, we **stay at the same place** when a duplicate is removed.

---

### 🧩 Step-by-Step Full Example

**Input:**  

head = 1 → 1 → 2 → 2 → 3

Initialize:
1 → 1 → 2 → 2 → 3
h
r

(`r` will remember the start of the list)

---

#### Step 1:
Duplicate found (`1 == 1`) → remove next:


1 → 2 → 2 → 3
h
r


---

#### Step 2:
No duplicate (`1 != 2`) → move head forward:

1 → 2 → 2 → 3
r h


---

#### Step 4:
No duplicate (`2 != 3`) → move head forward:

1 → 2 → 3
r h

---

#### Step 5:
End of list reached ✅

---

### ✅ Final Linked List:


1 → 2 → 3

Since `r` points to the start of the list,  
**return `r` as the result**.

---

### 💡 Summary

| Action | Condition | What to Do |
|:--------|:------------|:------------|
| Found duplicate | `head.val == head.next.val` | `head.next = head.next.next` (don’t move head) |
| No duplicate | `head.val != head.next.val` | `head = head.next` |
| After loop | — | `return res` (start node reference) |
